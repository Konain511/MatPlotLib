### IMPORTING

### SIMPLE GRAPH PLOT

```python
import numpy as np
```

```python
xpoints =np.array([0,6])
ypoints =np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints =np.array([-6,6])
ypoints =np.array([-250,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints =np.array([1,8])
ypoints =np.array([3,10])
plt.plot(xpoints,ypoints)
plt.show()
```

### PLOTTING WITH DIFFERENT VALUES 

```python
xpoints =np.array([4,5,6,7])
ypoints =np.array([8,9,2,1])
plt.plot(xpoints,ypoints)
plt.show()
```

### PLOTTING WITHOUT LINE

```python
xpoints =np.array([4,5,6,7,8])
ypoints =np.array([2,4,6,7,12])
plt.plot(xpoints,ypoints,'+')
plt.show()
```

### MULTIPLE POINTS

```python
xpoints =np.array([1,2,6,8])
ypoints =np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

### THREE POINTS GRAPH

```python
xpoints =np.array([1,2,6])
ypoints =np.array([3,8,1])
zpoints =np.array([4,5,6])
plt.plot(xpoints,ypoints,zpoints)
plt.show()
```

### DEFAULT 'x' POINT

```python
# xpoints =np.array([1,2,6,8])
ypoints =np.array([3,8,1,10])
plt.plot(ypoints)
plt.show()
```

### MARKERS AND MARKER-SIZE

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o',ms=30)
plt.show()
```

### FORMAT STRINGS fmt

```python
# Short syntax to show marjer ,line and color...
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o',color = "#33ccff",ms=30)
plt.show()
```

### MARKER EDGE COLOR

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o--y',ms=30,mec="b")
plt.show()
```

### LINE STYLE

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o',ms=30,mec="b",ls="dotted")
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o',ms=30,mec="b",ls="dashdot")
plt.show()
```

### LINE COLOR

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker='o',color='r',ms=30,mec="b",ls="dashdot")
plt.show()
```
